```python
def dfgen(s, dfx):
    np.random.seed(s)
    
    from sklearn.model_selection import train_test_split
    dfx_full_train, dfx_test = train_test_split(dfx, test_size=0.2, random_state=1)
    dfx_train,dfx_val=train_test_split(dfx_full_train, test_size=0.25, random_state=1)
    
    dfx_train=dfx_train.reset_index(drop=True)
    dfx_val=dfx_val.reset_index(drop=True)
    dfx_test=dfx_test.reset_index(drop=True)
    
    return dfx_train, dfx_val, dfx_test
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
def accuracy_calc(s, df1t, df1v):
    #Converting Dataframe into Dictionary
    small_train_dicts=df1t[s].to_dict(orient='records')
    small_val_dicts=df1v[s].to_dict(orient='records')

    #Converting Dictionaries into Vectorized Dictionaries
    dv_small=DictVectorizer(sparse=False)
    dv_small.fit(small_train_dicts) 

    #Preparing Training Data
    X_train_small= dv_small.transform(small_train_dicts)
    y_train=df1t.price.values

    #Preparing Validation Data
    X_val_small= dv_small.transform(small_val_dicts)
    y_val_small=df1v.price.values

    #Logistic regression
    model_small=LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
    model_small.fit(X_train_small, y_train)

    #Bias (w0) and Weights(w)
    w0=model_small.intercept_[0]
    w=model_small.coef_[0].round(3)

    #Weights as Dictionaries
    dict(zip(dv_small.get_feature_names_out(), model_small.coef_[0].round(3)))

    #Predicting y variable values
    y_pred_small=model_small.predict_proba(X_val_small)[:, 1]

    #Converting probabilities into binary
    churn_decision=(y_pred_small>=0.5)
    churn_decision.astype(int)

    #Accuracy
    accuracy=(y_val_small==churn_decision).mean()
    return accuracy
```

```python
var=['make', 'model', 'year', 'engine_hp', 'engine_cylinders', 'transmission_type', 'vehicle_style', 'highway_mpg', 'city_mpg']
acc=accuracy_calc(var, df1_train, df1_val)
acc
```
